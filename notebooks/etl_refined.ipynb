```python
import pandas as pd

# 1. Carregamento e Seleção
df = pd.read_csv('/home/alencaravelar/Desktop/zetta-lab/zetta-lab2/zetta-lab2/data/raw/dados_bh.csv')
target = 'IDHM'
features = [
    'T_ANALF15M', 'T_ATRASO_2_BASICO', 'T_FUND18M', 'AGUA_ESGOTO',
    'T_DENS', 'T_LIXO', 'GINI', 'PPOB', 'T_FUNDIN18MINF',
    'P_FORMAL', 'T_DES18M', 'RAZDEP'
]

# 2. Filtragem e Conversão Numérica
df_refined = df[features + [target]].copy()
df_refined = df_refined.apply(pd.to_numeric, errors='coerce')

# 3. Remoção de Nulos (os 18 registros identificados na EDA)
df_refined = df_refined.dropna(subset=[target] + features)

# 4. Filtro de Sanidade (Apenas valores positivos para taxas e índices)
for col in features:
    df_refined = df_refined[df_refined[col] >= 0]

# 5. Salvamento do Dataset pronto para o Modelo
df_refined.to_csv('/home/alencaravelar/Desktop/zetta-lab/zetta-lab2/zetta-lab2/data/refined/base_udh_refined.csv', index=False)
print(f"Limpeza concluída. Registros finais: {len(df_refined)}")
```
